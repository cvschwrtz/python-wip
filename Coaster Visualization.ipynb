```python
# Import internal library
import codecademylib3

# 1 
# Import necessary libraries
import pandas as pd
from matplotlib import pyplot as plt

# load rankings data
wood = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
#print(wood)

# load rankings data
steel = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')
#print(steel.head())

# 2
# Create a function to plot rankings over time for 1 roller coaster
def wood_rank (name, park):
  dfwood = wood[(wood['Name'] == name) & (wood['Park'] == park)]
  plt.plot(dfwood['Year of Rank'], dfwood['Rank'])
  plt.ylabel('Rank')
  plt.xlabel('Year')
  plt.title(name + ' Rank Over Time')
  plt.show()
  plt.clf()

# 3
# Create a plot of El Toro ranking over time
#wood_rank('El Toro', 'Six Flags Great Adventure')

# Create a plot of El Toro and Boulder dash hurricanes
def two_wood_rank(name1, name2, park1, park2):
  dfwood1 = wood[(wood['Name'] == name1) & (wood['Park'] == park1)]
  dfwood2 = wood[(wood['Name'] == name2) & (wood['Park'] == park2)]
  plt.plot(dfwood1['Year of Rank'], dfwood1['Rank'], label=name1, color='red', marker='o')
  plt.plot(dfwood2['Year of Rank'], dfwood2['Rank'], label=name2, color='blue', marker='o')
  plt.ylabel('Rank')
  plt.xlabel('Year')
  plt.legend()
  plt.title('Rank Over Time')
  plt.show()
  plt.clf()

#two_wood_rank('El Toro', 'Boulder Dash', 'Six Flags Great Adventure', 'Lake Compounce')

# 4
# Create a function to plot top n rankings over time
def top_rank(df, n):
  top = df[df['Rank'] <= n]
  fig, ax = plt.subplots(figsize=(10, 10))
  for coaster in set(top['Name']):
    coaster_rankings = top[top['Name'] == coaster]
    ax.plot(coaster_rankings['Year of Rank'], coaster_rankings['Rank'], label=coaster, marker='o')
  ax.set_yticks([i for i in range(1, 6)])
  plt.xlabel('Year')
  plt.ylabel('Rank')
  plt.title('Top Rankings')
  plt.legend()
  plt.show()
  plt.clf()

# Create a plot of top n rankings over time
#top_rank(wood, 5)

# 5
# load roller coaster data
coaster_df = pd.read_csv('roller_coasters.csv')
print(coaster_df)

# 6
# Create a function to plot histogram of column values
def num_hist(df, column):
  plt.hist(df[column].dropna())
  plt.title('Histogram of Roller Coaster {}'.format(column))
  plt.xlabel(column)
  plt.ylabel('Count')
  plt.show()
  plt.clf()

# Create histogram of roller coaster speed
#num_hist(coaster_df, 'speed')

# Create histogram of roller coaster length
#num_hist(coaster_df, 'length')

# Create histogram of roller coaster number of inversions
#num_hist(coaster_df, 'num_inversions')

# Create a function to plot histogram of height values
def height_hist(df):
  heights = df[df['height'] <= 140]['height'].dropna()
  plt.hist(heights)
  plt.title('Histogram of Roller Coaster Height')
  plt.xlabel('Height')
  plt.ylabel('Count')
  plt.show()
  plt.clf()

# Create a histogram of roller coaster height
#height_hist(coaster_df)

# 7
# Create a function to plot inversions by coaster at park
def num_inv(df, park):
  park_coasters = df[df['park'] == park]
  park_coasters = park_coasters.sort_values('num_inversions', ascending=False)
  coasters = park_coasters['name']
  number_inv = park_coasters['num_inversions']
  plt.bar(range(len(number_inv)), number_inv)
  ax = plt.subplot()
  ax.set_xticks(range(len(coasters)))
  ax.set_xticklabels(coasters, rotation=90)
  plt.title('Number of Inversions Per Coaster at {}'.format(park))
  plt.xlabel('Roller Coaster')
  plt.ylabel('# of Inversions')
  plt.tight_layout()
  plt.show()
  plt.clf()

# Create barplot of inversions by roller coasters
#num_inv(coaster_df, 'Six Flags Great Adventure')

# 8
# Create a function to plot a pie chart of status.operating
def pie_chart(df):
  operating = df[df['status'] == 'status.operating']
  closed = df[df['status'] == 'status.closed.definitely']
  num_operating = len(operating)
  num_closed = len(closed)
  status_counts = [num_operating, num_closed]
  plt.pie(status_counts, autopct='%0.1f%%')
  plt.axis('equal')
  plt.title('Roller Coaster Status')
  plt.legend(['Operating', 'Closed'])
  plt.show()
  plt.clf()

# Create pie chart of roller coasters
#pie_chart(coaster_df)

# 9
# Create a function to plot scatter of any two columns
def scatter_coaster(df, column1, column2):
  plt.scatter(df[column1], df[column2])
  plt.title('Scatter Plot of {} vs. {}'.format(column2, column1))
  plt.xlabel(column1)
  plt.ylabel(column2)
  plt.show()
  plt.clf()

#scatter_coaster(coaster_df, 'length', 'speed')

# Create a function to plot scatter of speed vs height
def scatter_height_speed(df):
  df = df[df['height'] < 140]
  plt.scatter(df['height'], df['speed'])
  plt.title('Scatter Plot of Speed vs. Height')
  plt.xlabel('Height')
  plt.ylabel('Speed')
  plt.show()
  plt.clf()

# Create a scatter plot of roller coaster height by speed

#scatter_height_speed(coaster_df)
```
