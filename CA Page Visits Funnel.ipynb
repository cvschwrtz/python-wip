```python
import codecademylib
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])

#print(visits.head(10))
#print(cart.head(10))
#print(checkout.head(10))
#print(purchase.head(10))

visits_cart = pd.merge(visits, cart, how = 'left')
#print(visits_cart)

visits_cart_row = len(visits_cart)
#print(visits_cart_row)

is_null = len(visits_cart[visits_cart.cart_time.isnull()])

no_cart = (float(is_null) / float(visits_cart_row) * 100)
#print(no_cart)

cart_check = pd.merge(cart, checkout, how='left')
cart_check_len = len(cart_check)
check_is_null = len(cart_check[cart_check.checkout_time.isnull()])
no_check = (float(check_is_null) / float(cart_check_len) * 100)
#print(no_check)

all_data = visits.merge(cart, how='left')\
                 .merge(checkout, how='left')\
                 .merge(purchase, how='left')
#print(all_data.head(20))

check_purchase = pd.merge(checkout, purchase, how='left')
check_purchase_len = len(check_purchase)
purchase_is_null = len(check_purchase[check_purchase.purchase_time.isnull()])
no_purchase = (float(purchase_is_null) / float(check_purchase_len) * 100)
#print(no_purchase)

all_data['time_to_purchase'] = \
    all_data.purchase_time - \
    all_data.visit_time
#print(all_data)
#print(all_data.time_to_purchase)
#print(all_data.time_to_purchase.mean())
```
